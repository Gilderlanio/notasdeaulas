```python
# Verificar isomorfismo entre grafos

# Criar dois grafos com estruturas similares
C1 = nx.Graph()
C2 = nx.Graph()

# Adicionar as arestas de C1
C1.add_edge(1, 2)
C1.add_edge(1, 3)
C1.add_edge(1, 4)
C1.add_edge(1, 5)
C1.add_edge(1, 6)
C1.add_edge(2, 5)
C1.add_edge(3, 4)
C1.add_edge(3, 5)
C1.add_edge(4, 6)
C1.add_edge(5, 6)

# Adicionar as arestas de C2
C2.add_edge("A", "B")
C2.add_edge("A", "C")
C2.add_edge("A", "D")
C2.add_edge("A", "E")
C2.add_edge("A", "F")
C2.add_edge("B", "E")
C2.add_edge("C", "D")
C2.add_edge("C", "E")
C2.add_edge("D", "F")
C2.add_edge("E", "F")

nx.draw(C1)

nx.is_isomorphic(C1, C2) # Verificar isomorfismo entre grafos
```

```python
# Exemplo de identificação de pontes

import networkx as nx
I = nx.Graph() 
I.add_edges_from([(1, 2),(1, 5),(1,4),(1,6),(5,2),(5,3),(3,4),(6,9),(6,10),(6,8),(9,10),(9,8),(8,10),(8,7),(10,7)])
nx.draw(I)
```

```python
nx.has_bridges(I) # Informar se o grafo possui pontes
```

```python
list(nx.bridges(I)) # Localizar pontes do grafo
```

```python
print(nx.clustering(I, 6)) # Calcular o coeficiencte de agrupamento do nó 6 do grafo
print(nx.clustering(I, 1)) # Calcular o coeficiencte de agrupamento do nó 1 do grafo
```

```python
# Utilização do algoritmo de Girvan Newman para separar um grafo em comunidades

from networkx.algorithms.community.centrality import girvan_newman
components = girvan_newman(I)

components_list = list(components)
#print(components_list) # Visualizar a estrutura dos componentes

# Separar os componentes
components_1_division = [sorted(comp) for comp in components_list[0]]
#components_2_division = [sorted(comp) for comp in components_list[1]]
#components_3_division = [sorted(comp) for comp in components_list[2]]

print(components_1_division)
#print(components_2_division)
#print(components_3_division)
```

```python
# Criando subgrafos para demonstrar o uso do algoritmo Girven Newman

components_I = [I.subgraph(comp) for comp in components_1_division] # Receber uma lista com subgrafos de I

print(len(components_I)) # Mostrar a quantidade de comunidades esperadas

I1, I2 = components_I # Separar os subgrafos em 2 váriaveis

# Mostrar o impacto da remoção de uma ponte no coeficiente de agrupamento
print(nx.clustering(I2, 6))
print(nx.clustering(I1, 1))


import matplotlib.pyplot as plt # Importar parar imprimir 2 ou mais figuras em uma mesma célula

plt.figure(1) # Iniciar bloco para figura 1
nx.draw(I1) # Modelar o Grafo I1

plt.figure(2) # Iniciar bloco para figura 2
nx.draw(I2) # Modelar o grafo I2

plt.show() # Finalizar e plotar as figuras
```

```python
# Exemplo de utilizando de grafo multi-aresta para construção de grafos

J = nx.MultiGraph()

J.add_edge(1, 2, Age = 22)
J.add_edge(1, 2, Sex = "M")
J.add_edge(1, 2, Type = "case")

J.add_edge(1, 3, Sex = "F")
J.add_edge(1, 3, Type = "case")

J.add_edge(2, 3, Age = 27)
J.add_edge(2, 3, Type = "control")

#print(J.edges(data = True))

pos_J = nx.circular_layout(J)
nx.draw_networkx_nodes(J, pos_J)
nx.draw_networkx_edges(J, pos_J)
nx.draw_networkx_labels(J, pos_J)
```

```python
# Retirar arestas que contem um atributo definido para construção de subgrafos que podem ser modelados

atribute = "Sex" # Receber um atributo de dentro do Grafo "J"

edges = nx.get_edge_attributes(J, atribute) # Obter somente grafos associados com atributo definido
#print(edges)

edge_list = [(key[0], key[1], {atribute:edges[key]}) for key in edges] # Criar uma lista com informações para construção de um novo grafo
# a lista conterá o nó de origem, nó de destino e um dicionário com chave e valores associadas a váriavel "atribute"

#print(edge_list)

A = nx.Graph() # Criar um novo objeto "grafo"
A.add_edges_from(edge_list) # Receber arestas, baseado na lista "edge_list" criada

edge_labels = nx.get_edge_attributes(A, atribute) # Associar uma aresta com valor de "atribute" em grafo "A"
#print(edge_labels)

# Desenhar o subgrafo
pos = nx.circular_layout(A)

nx.draw_networkx_nodes(A, pos)
nx.draw_networkx_edges(A, pos)
nx.draw_networkx_labels(A, pos)
nx.draw_networkx_edge_labels(A, pos, edge_labels)
```

```python
# Layout de grafo bipartido
from networkx.drawing.layout import bipartite_layout

# Criar os conjuntos do grafo bipartido
U = ["A", "B", "C"]
V = [1, 2, 3, 4]

B = nx.DiGraph() # Criar um objeto "grafo" que será direcionado

# Montar o grafo
B.add_nodes_from(U)
B.add_nodes_from(V)
B.add_edges_from([("A", 1), ("A", 4), ("B", 1), ("B", 2), ("C", 3), ("C", 4)])

pos = bipartite_layout(B, U) # Definir a posição dos nós no grafo

# Modelar os grafos
#nx.draw(B, with_labels = True)
nx.draw_networkx(B, pos)
```
